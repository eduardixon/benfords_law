```python
import matplotlib.pyplot as plt
```

```python
poblaciones = []
with open("worldcities.csv", "r", encoding = "utf8") as f:
    for line in f:
        new_list = line.split(",")
        poblaciones.append((new_list[-2])[1:-1])
poblaciones = poblaciones[1:]
print(len(poblaciones))
```

```python
first_digit = [int(item[0]) for item in poblaciones if len(item) > 0]
print(first_digit)
```

```python
x = [digit for digit in range(1, 10)]
y = [first_digit.count(element) for element in x]
print(y)
```

```python
list_x = [digits for digits in range(1, 10)]
list_y = [0.301, 0.176, 0.125, 0.097, 0.079, 0.067, 0.058, 0.058, 0.046]
```

```python
plt.bar(x, y, color = "r", label = "Results")
plt.title("Benfords law practice")
plt.ylabel("Quantity of countrys")
plt.xlabel("Number it starts with")
plt.legend()
plt.show()
```

```python
total = 0
for element in y:
    total += element
print(total)    
```

```python
list_y_ratios =[element/total for element in y]
print(list_x)
```

```python
plt.title("Benfords law ratios")
plt.bar(list_x, list_y_ratios, color = "red", label = "Results")
plt.plot(list_x, list_y, color = "green", label = "Expectations", marker = ".")
plt.legend()
```
